```python
import pandas as pd
from lxml import etree
import requests
import os
import sys
import time
```

```python
sys.path.append(os.path.abspath("./../scripts/"))
```

```python
import parse_sexual_anotation
```

```python
wdir_str = "./../../../sexual-annotation/"
```

```python
taxonomy_file_str = "taxonomy.xml"
```

```python
taxonomy_el = parse_sexual_anotation.open_taxonomy(wdir_str, taxonomy_file_str)

```

```python
namespaces_dict = {'tei':'http://www.tei-c.org/ns/1.0','xi':'http://www.w3.org/2001/XInclude'}
```

```python
results = []
for category in taxonomy_el.xpath('//tei:category', namespaces = namespaces_dict):
    id_ = category.xpath("./@xml:id", namespaces = namespaces_dict)[0]
    print(id_)
    type_ = category.xpath("./tei:desc/@type", namespaces = namespaces_dict)[0]
    text = category.xpath("./tei:desc/tei:gloss/text()", namespaces = namespaces_dict)[0]
    try:
        main_category = category.xpath("./ancestor::tei:category[./parent::tei:taxonomy]/@xml:id", namespaces = namespaces_dict)[0]
    except:
        main_category = id_
    results.append([id_, type_, main_category, text])
```

```python
results[0:2]
```

```python
df = pd.DataFrame(results, columns = ["category", "explicitness", "main_type", "description"])
```

```python
df.to_csv("./taxonomy.tsv", sep="\t")
```
